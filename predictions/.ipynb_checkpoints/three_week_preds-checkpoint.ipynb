```python
days = ['day_' + str(i) for i in range(21)]

def pipeline(data):
    data = data.pipe(interpolate_vals, days)
    data = data.pipe(log_apply, days + ['days_since_release', 'Next Full Week Streams'], prediction_col='Next Full Week Streams')
    return data

train = pipeline(train)
train
```
